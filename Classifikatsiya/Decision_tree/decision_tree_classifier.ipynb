```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
data = {
    'Yosh': [18, 25, 30, 22, 40, 35, 28, 50],
    'Ishlaydi': [0, 1, 1, 0, 1, 1, 1, 1],
    'Boyi': [170, 165, 180, 175, 160, 155, 185, 150],
    'Sport': [1, 1, 1, 1, 0, 0, 1, 0]
}
```

```python
df = pd.DataFrame(data)
```

```python
X = df[['Yosh', 'Ishlaydi', 'Boyi']]
y = df['Sport']
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
plt.figure(figsize=(10,6))
tree.plot_tree(model, feature_names=X.columns, class_names=["Yo'q", "Ha"], filled=True)
plt.show()
```

```python
yangi_odam = pd.DataFrame([[26, 1, 172]], columns=['Yosh', 'Ishlaydi', 'Boyi'])
bashorat = model.predict(yangi_odam)
print("Sport bilan shug‘ullanadimi?", "Ha" if bashorat[0] == 1 else "Yo'q")
```
